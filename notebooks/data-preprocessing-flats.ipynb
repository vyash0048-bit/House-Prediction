```python
def treat_price(x):
    if not isinstance(x, list) or len(x) < 2:
        return None      # or np.nan

    if x[1] == 'Lac':
        return round(float(x[0]) / 100, 2)

    elif x[1] == 'Cr':
        return float(x[0])

    else:
        return None
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```
